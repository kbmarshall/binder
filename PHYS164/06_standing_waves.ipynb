If one end of a string is forced to oscillate, the oscillations will generate a wave which will travel the length of the string.  This wave moves with speed $v$ and amplitude $A$ along the string.

The relationship between wave speed, frequency, and wavelength is given by:

$$ v = f \lambda $$

If the string is fixed at both ends, at certain frequencies a *standing wave* can be produced.  Standing waves result from the interference of the original wave with its reflection.  When the frequency is just right, there will be locations along the string (called nodes) where the original wave and its reflection always cancel each other.  At these points, the string will remain stationary.  There will also be points (called anti-nodes) where the reflection adds with the original wave to produce an oscillation with double the amplitude.

The distance between nodes (or anti-nodes) is always half a wavelength.  Thus by finding the frequencies where standing waves occur, and measuring the wavelength, you can calculate the speed of the wave.

The wave speed can also be found from the tension in the wave $T$ and the mass density $\mu$, which is the mass per length of the string.

$$ v = \sqrt{\frac{T}{\mu}} $$

As long as $T$ and $\mu$ are kept constant, the wave speed should remain the same.

In order for a standing wave to occur, the distance between the ends of the string $L$ must be a multiple of half the wavelength, or $L = m \lambda / 2$, where $m=1,2,3,...$  That means the frequencies where you find standing waves are given by:

$$ f = \frac{m v}{2L} = m f_0 $$

where $f_0$ is known as the fundamental frequency, and all other frequencies are multiples of this fundamental.

As a reminder, if you want to fit data with a straight line and get uncertainties, the procedure is:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)              # Generating fake data here
deltay = np.random.randn(5)   # Use your own data
y = 2*x+np.random.randn(5)    # Use your own data and uncertainties

p,pcov = np.polyfit(x,y,1,cov=True)
print(p)
```

The highest power of x is printed first, so in this case our data would have a slope near 2.

To get the uncertainties in our fit parameters, we take the diagonals of the covariance matrix and square them:

```python
np.diag(pcov)**2
```

Let's plot our data with the fit:

```python
f = np.poly1d(p)

plt.errorbar(x,y,yerr=deltay,fmt='o')
plt.plot(x,f(x),'-')
plt.show()
```
